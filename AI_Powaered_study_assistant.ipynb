Step.2) Import module from gemini Packege

```python
from google import genai
from google.colab import userdata
```

Step.3) Use Client class to create an instance(for interacting with Gemini API)

```python
client = genai.Client(api_key =userdata.get('GEMINI_API_KEY'))
```

```python
print(client)
```

Step.4) Use client object for communicating  to Gemini LLM

```python
def study_assistant(user_prompt):

  response = client.models.generate_content(
      model = "gemini-2.5-flash",
      contents = user_prompt
      )
  return response
```

```python
output = study_assistant("Explain gen ai")
print(output.text)
```
